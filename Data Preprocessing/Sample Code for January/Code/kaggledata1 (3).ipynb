*Import pandas library and read csv file and assing it into dataframe.*

```python
import pandas as pd
df = pd.read_csv('T1.csv',parse_dates=['Date/Time'])
df.head()
```

*Rename the coloumn.*

```python
df.rename(columns = {'Date/Time':'Date'}, inplace = True) 
```

```python
df.head()
```

```python
type('Date[0]')
```

*Set index as Date coloumn.*

```python
df.set_index('Date',inplace=True)
df.head()
```

*To fill the data make full list of day in year of 2018 at frequency of 10 min and reindex the dataframe.*

```python
dt=pd.date_range("01-01-2018","01-01-2019",freq='10min')
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

*Fill nan values with mean of coloumns.*

```python
new_df1=df.fillna({'LV ActivePower (kW)':1307.684332,'Wind Speed (m/s)':7.557952})
```

*Remove the unnessecary coloumns.*

```python
new_df1.drop('Theoretical_Power_Curve (KWh)',axis=1,inplace=True)
```

```python
new_df1.drop('Wind Direction (°)',axis=1,inplace=True)
```
