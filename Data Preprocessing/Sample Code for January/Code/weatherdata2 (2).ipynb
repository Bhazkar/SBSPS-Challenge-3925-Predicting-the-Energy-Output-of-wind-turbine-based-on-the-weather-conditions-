*Import pandas library and read csv file and assigning into a dataframe.*

```python
import pandas as pd
df = pd.read_csv('january1.csv',parse_dates=['Date'])
df.head()
```

*Set index as Date coloumn.*

```python
df.set_index('Date',inplace=True)
df.head()
```

*Remove the unnessecary coloumns from dataframe.*

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.drop('WindSpeed',axis=1,inplace=True)
```
