```python
import pandas as pd
import numpy as np
df = pd.read_csv('janfinal.csv',parse_dates=['Date'])
df.head()
```

```python
import matplotlib.pyplot as plt
plt.plot(df['Wind Speed (miles/hr)'],df['Active Power (kw)'],'o')
plt.title('Active Power(kw) vs Wind Speed(miles/hr)',fontsize=18)
plt.xlabel('Wind Speed(miles/hr)',fontsize=14)
plt.ylabel('Active Power(kw)',fontsize=14)
```

```python
plt.plot(df['Tempreature'],df['Active Power (kw)'],'o')
plt.title('Active Power(kw) vs Tempreature',fontsize=18)
plt.xlabel('Tempreature',fontsize=14)
plt.ylabel('Active Power(kw)',fontsize=14)
```

```python
plt.plot(df['Dp'],df['Active Power (kw)'],'o') #dp =dew point
plt.title('Active Power(kw) vs Dp',fontsize=18)
plt.xlabel('Dp',fontsize=14)
plt.ylabel('Active Power(kw)',fontsize=14)
```

```python
plt.plot(df['Humidity'],df['Active Power (kw)'],'o')
plt.title('Active Power(kw) vs Humidity',fontsize=18)
plt.xlabel('Humidity',fontsize=14)
plt.ylabel('Active Power(kw)',fontsize=14)
```
