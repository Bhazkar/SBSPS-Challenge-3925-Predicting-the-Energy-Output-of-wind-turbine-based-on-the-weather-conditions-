```python
def convert24(str1): 
      
    # Checking if last two elements of time 
    # is AM and first two elements are 12 
    if str1[-2:] == "AM" and str1[:2] == "12": 
        return "00" + str1[2:-3] 
          
    # remove the AM     
    elif str1[-2:] == "AM": 
        return str1[:-3] 
      
    # Checking if last two elements of time 
    # is PM and first two elements are 12    
    elif str1[-2:] == "PM" and str1[:2] == "12": 
        return str1[:-3] 
          
    else: 
          
        # add 12 to hours and remove PM 
        return str(int(str1.split(':')[0]) + 12) +':'+str1.split(':')[1][0:2]
```

```python
df1['DateTime4']=df1['Time'].apply(lambda x: x.split('.')[0]+':'+x.split('.')[1])
```

```python
df1['DateTime4']=df1['DateTime4'].apply(lambda x:convert24(x))
```

```python
df1.head()
```

```python
df1['DateTime4'] = pd.to_datetime(df1['DateTime4']).dt.time
```

```python
df1['DateTime4']=df1['DateTime4'].apply(lambda x:str(x))
```

```python
x='00-02-2018'
for i in range(len(df1)):
    if df1['DateTime4'].iloc[i]=='00:20:00':
        df1['DateTime4'].iloc[i]=str(int(x.split('-')[0])+1)+'-'+'-'.join(x.split('-')[1:])+' '+df1['DateTime4'].iloc[i]
        x=str(int(x.split('-')[0])+1)+'-'+'-'.join(x.split('-')[1:])
    else:
        df1['DateTime4'].iloc[i]=x+' '+df1['DateTime4'].iloc[i]
```
